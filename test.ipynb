```python
from pathlib import Path
import pandas as pd
entries = Path('dfs/')
dfs =[]
for entry in entries.iterdir():
    try:
        dfs.append(pd.read_csv(entry))
    except:
        'None'

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Jugadores'],
                                            how='outer'), dfs).fillna(0)
```

```python
df = pd.DataFrame({'Jugadores':Jugadores,
                   'Goles':Goles,
                   'Equipos':Equipos,
                   'MVP':MVP,
                   'WL':WL})
```
